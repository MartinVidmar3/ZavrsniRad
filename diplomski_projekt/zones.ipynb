```python
import pandas as pd
from pathlib import Path

floor_ids = [23]
years = [2018, 2019, 2020, 2021]
all_data = []

for year in years:
    for floor_id in floor_ids:
        path = f"../../results/{year}/{floor_id}/zone_stac_stanja_{floor_id}_{year}.csv"
        
        if Path(path).exists():
            df = pd.read_csv(path)
            all_data.append(df)
        else:
            print(f"File not found: {path}")

if all_data:
    combined_data = pd.concat(all_data, ignore_index=True)
    print("Successfully combined data. Shape:", combined_data.shape)
    combined_data.to_csv("../../results/cooling/23/zones.csv", index=False)
    print("Data written to 'results/fcu_type_id1.csv'")
else:
    print("No files were found.")
```
