```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
file_path = "../../../../results/2018/24/zones_24_year_2018filtriranje.csv"
file_path1 = "../../../../results/2018/24/15min/calorimeter_24_year_2018presjeci.csv"


data = pd.read_csv(file_path, index_col=[0])
data1=pd.read_csv(file_path1, index_col=[0])
```

```python
# Print the first two rows of each DataFrame
print("First two rows of data (filtriranje):")
print(data.head(2))

print("\nFirst two rows of data1 (presjeci):")
print(data1.head(2))
print("Number of rows:", len(data1))
```

```python
# Reset index so intersection_end becomes a column
data1 = data1.reset_index()

# Now intersection_end is a proper column
data1['intersection_end'] = pd.to_datetime(data1['intersection_end']).dt.round('min')

# Also round timestamp in data
data['timestamp'] = pd.to_datetime(data['timestamp']).dt.round('min')

# Merge on the rounded timestamps
merged = pd.merge(data1, data, left_on='intersection_end', right_on='timestamp', how='inner')

print(f"Number of merged rows: {len(merged)}")
print(merged.head())
zone_ids = [195, 196, 197, 198, 199, 200, 201, 202, 203, 204, 205, 206, 207]

# Count and print rows for each zone_id
for zone_id in zone_ids:
    count = len(merged[merged['zone_id'] == zone_id])
    print(f"zone_id {zone_id}: {count} rows")

merged.to_csv( '../../../../results/2018/24/15min/zone_stac_stanja_24_2018.csv ', index=False)

```
