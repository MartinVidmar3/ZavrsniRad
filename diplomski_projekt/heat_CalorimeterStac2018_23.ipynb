```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
file_path = "../../../results/2020/23/calorimeter_23_year_2020filtriranje_heating.csv"
file_path1 = "../../../results/2020/23/calorimeter_23_year_2020presjeci.csv"


data = pd.read_csv(file_path, index_col=[0])
data1=pd.read_csv(file_path1, index_col=[0])
print(data.columns)
```

```python
print("First two rows of data (filtriranje):")
print(data.head(2))

print("\nFirst two rows of data1 (presjeci):")
print(data1.head(2))
```

```python
data['timestamp'] = pd.to_datetime(data['timestamp'])
print(f"Broj redaka caorimrter.csv prije filtracije: {len(data)}")
print(f"Broj redaka presjeci.csv: {len(data1)}")

data1.index = pd.to_datetime(data1.index).floor('min')
data1['intersection_end'] = pd.to_datetime(data1['intersection_end']).dt.floor('min')  # Changed to .dt.floor

data['timestamp_rounded'] = data['timestamp'].dt.floor('min')

filtered_data = data[data['timestamp_rounded'].isin(data1['intersection_end'])]

filtered_data = filtered_data.drop(columns=['timestamp_rounded'])

print(f"Broj redaka caorimrter.csv poslije filtracije: {len(filtered_data)}")
print(filtered_data.head())

filtered_data.to_csv('../../../results/2020/23/stac_stanja_23_2020_heating.csv', index=False)
```
