```python
calorim = pd.read_csv("../../../../../results/calorimeter_filtered24.csv") 
zones = pd.read_csv("../../../../../results/usporedbaThermal/24/speed1/30min/zones.csv") 
fcu_files = pd.read_csv("../../../../../results/24/fcu_type_id1_with_zone.csv") 
fcu_files1 = pd.read_csv("../../../../../results/24/fcu_type_id2_with_zone.csv") 
all_fcu_files = pd.concat([fcu_files, fcu_files1], ignore_index=True)
mapping = pd.read_csv("../../../../../results/zone_find.csv")

# Convert timestamps for all datasets
calorim['timestamp'] = pd.to_datetime(calorim['timestamp']).dt.floor('min')
zones['timestamp'] = pd.to_datetime(zones['timestamp']).dt.floor('min')
fcu_files['timestamp'] = pd.to_datetime(fcu_files['timestamp']).dt.floor('min')
fcu_files1['timestamp'] = pd.to_datetime(fcu_files1['timestamp']).dt.floor('min')

# Combine fcu_files
all_fcu_files = pd.concat([fcu_files, fcu_files1], ignore_index=True)

# Merge zones and calorimeter
merged = pd.merge(zones, calorim, on='timestamp', how='inner')

# Merge with all_fcu_files
final_merged = pd.merge(merged, all_fcu_files, on=['timestamp', 'zone_id'], how='inner')

# Sort by timestamp
final_merged = final_merged.sort_values('timestamp')

# Save to CSV
final_merged.to_csv("../../../../../results/usporedbaThermal/24/speed1/30min/merged_data.csv", index=False)

print(final_merged.head())
```

```python
final_merged = pd.read_csv("../../../../../results/usporedbaThermal/24/speed1/30min/merged_data.csv")

CP_WATER = 4184  # J/(kg·K)

# Convert from L/h to kg/s (1 L ≈ 1 kg)
scale_factor = 0.1  
final_merged['mass_flow_kg_s'] = (final_merged['mass_volume_flow'] * scale_factor) / 3600

# ΔT
final_merged['delta_T'] = final_merged['supply_temperature'] - final_merged['return_medium_temperature']

# Thermal power in kW
final_merged['calculated_thermal_power'] = (
    final_merged['mass_flow_kg_s'] * CP_WATER * final_merged['delta_T'] / 1000
)

print(final_merged[['thermal_power', 'calculated_thermal_power', 'mass_flow_kg_s', 'delta_T']].head(20))


final_merged.to_csv("../../../../../results/fcu_type_id2/23/speed1/30min/calc_thermal.csv", index=False)
print("Data written to ../../../../results/fcu_type_id1/speed0/15min/merged_data.cs")
```

```python
import pandas as pd


zone_ids = [195, 196, 197, 198, 199, 201, 202, 204, 205, 206, 207, 200, 203]

# Filter for zone_ids of interest
df_filtered = final_merged[final_merged['zone_id'].isin(zone_ids)].copy()

# Convert timestamp to datetime
df_filtered['timestamp'] = pd.to_datetime(df_filtered['timestamp'])

# Add minute-level timestamp
df_filtered['minute'] = df_filtered['timestamp'].dt.floor('min')

# Step 1: Ensure each zone_id only counted once per minute
df_unique = (
    df_filtered
    .drop_duplicates(subset=['minute', 'zone_id'])
)

# Step 2: Group by minute
grouped = df_unique.groupby('minute')

# Step 3: Keep only minutes with all zone_ids present
complete_minutes = grouped.filter(lambda g: set(g['zone_id']) == set(zone_ids))

# Step 4: Sum calculated_thermal_power for complete minutes
result = (
    complete_minutes.groupby('minute')['calculated_thermal_power']
    .sum()
    .reset_index()
    .rename(columns={'calculated_thermal_power': 'total_thermal'})
)

# Step 5: Save to CSV
result.to_csv('total_thermal_per_minute.csv', index=False)

print(result.head())

```
