```python
import pandas as pd
from pathlib import Path

floor_id = 23
years = [2018, 2019, 2020, 2021, 2022]
interval = "15min"
all_data = []


total_rows=0
for year in years:
    path = f"../../results/{year}/{floor_id}/stac_stanja_{floor_id}_{year}_cooling.csv"
    
    if Path(path).exists():
        df = pd.read_csv(path)
        print(df.head())
        print(f"Year {year}: Original data shape - {df.shape}")
        total_rows += len(df)
        all_data.append(df)
    else:
        print(f"File not found: {path}")
print(f"Total number of rows across all years: {total_rows}")
if all_data:
    combined_data = pd.concat(all_data, ignore_index=True)
    print("Successfully combined data. Shape:", combined_data.shape)
    combined_data.to_csv("../../results/cooling/23/sveGodine.csv", index=False)
else:
    print("No files were found.")
print(combined_data.head())
```
